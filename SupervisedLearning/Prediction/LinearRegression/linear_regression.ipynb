```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("data1.csv",sep=";")
df.head(3)
```

```python
X = df.experience.values.reshape(-1,1)
y = df.salary.values.reshape(-1,1)
```

```python
plt.scatter(X,y)
plt.xlabel("experience-year")
plt.ylabel("salary-tl")
plt.title("experience-salary relationship")
plt.grid(True)
plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
y_intercept = np.array([0]).reshape(1,-1)
b0 = model.predict(y_intercept)

b0_ = model.intercept_
print("b0_: ",b0_)

b1 = model.coef_
print("b1: ",b1)
```

```python
# salary = 1663 + 11338 * experience
experience = 11

new_salary = model.predict(np.array([experience]).reshape(1,-1))
print("The salary of person with {} years of experience is {} TL".format(experience,new_salary[0]))
```
