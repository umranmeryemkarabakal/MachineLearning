```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
df = pd.read_csv("data1.csv",sep=";")
df.head(3)
```

```python
X = df.car_price.values.reshape(-1,1)
y = df.car_max_speed.values.reshape(-1,1)
```

```python
plt.scatter(X,y)
plt.ylabel("maximum speed of the car")
plt.xlabel("price of the car")
plt.title("speed-price relationship")
plt.grid(True)
plt.show()
```

```python
#polynomial regression => y = b0 + b1*x + b2*x^2 + ... + bn*x^n
```

```python
model = PolynomialFeatures(degree = 4)
```

```python
X_polynomial = model.fit_transform(X)
#output => b0 + b1*x + b2*x^2 + b3*x^3 + b4*x^4
```

```python
model1 = LinearRegression()
```

```python
model1.fit(X_polynomial,y)
```

```python
y_predict = model1.predict(X_polynomial)
```

```python
plt.scatter(X,y)
plt.plot(X,y_predict)
plt.ylabel("maximum speed of the car")
plt.xlabel("price of the car")
plt.title("speed-price relationship")
plt.grid(True)
plt.show()
```

```python
y_pred_poly = model1.predict(X_polynomial)
```

```python
r2_poly = r2_score(y, y_pred_poly)
print(f"Polynomial R² Skoru: {r2_poly}")
```

```python
mse_poly = mean_squared_error(y,y_pred_poly)
print(f"Polynomial Mean Squared Error (MSE): {mse_poly}")
```

```python
rmse_poly = np.sqrt(mse_poly)
print(f"Polynomial Root Mean Squared Error (RMSE): {rmse_poly}")
```

```python
mae_poly = mean_absolute_error(y,y_pred_poly)
print(f"Polynomial Mean Absolute Error (MAE): {mae_poly}")
```
