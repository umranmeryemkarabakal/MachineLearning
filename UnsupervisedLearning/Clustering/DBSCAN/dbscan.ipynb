```python
# yoğunluğa dayalı kümeleme
# gürültülü kümelerin yoğunluk tabanlı mekansal kümelemesidir
# önce benzerlikleri hesaplar sonra kümeleme yapar
# epsilon, bir veri noktasının etrafındaki komşuluk
# epsilon çok küçük seçilirse aykırı değerler artar, çok büyük seçilirse kümeleme yetersiz olur
# minPts: epsilon yarıçap içindeki minimum komşu sayısı
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X, _ = make_moons(n_samples=200, noise = 0.1, random_state=42)
```

```python
dbscan_model = DBSCAN(eps=0.2, min_samples=5)
# eps: Bir veri noktasının etrafındaki komşuluk yarıçapı
# min_samples: Epsilon yarıçapı içindeki minimum komşu sayısı
```

```python
cluster = dbscan_model.fit_predict(X)
```

```python
n_cluster = len(set(cluster)) - (1 if -1 in cluster else 0)
n_noise = list(cluster).count(-1)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[cluster==-1,0],X[cluster == -1,1],c="gray",marker = "o",s=30,label = "noise") 

for i in range(n_cluster):
    plt.scatter(X[cluster == i,0],X[cluster == i,1],label = f"cluster {i+1}")

plt.title("DBSCAN clustering")
plt.xlabel("features 1")
plt.ylabel("features 2")
plt.legend()
plt.show()
```
